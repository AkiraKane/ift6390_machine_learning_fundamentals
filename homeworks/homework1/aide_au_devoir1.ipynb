## Charger les données avec numpy

```python
# Pour charger les données, on doit spécifier le "delimiter", dans ce cas ci ","
train_set = np.loadtxt("train_images.txt", delimiter=",")
train_labels = np.loadtxt("train_labels.txt", delimiter=",")

print "X est de dimension"
print train_set.shape, "\n"
print "Y est de dimension"
print train_labels.shape, "\n"
```

## Calculer efficacement le taux d'erreur de classification

```python
N = train_set.shape[0]
c = train_labels.shape[1]

# Soit une matrice bidon qui représente les prédictions d'un modèle
y_hat = np.random.multinomial(10, [1./float(c)]*c, size=N)

# On prend l'indice maximal de chaque rangée avec argmax(Y, axis=1)
predictions = np.argmax(y_hat, axis=1)

print "Les 10 premières prédictions"
print predictions[:10], "\n"

# On fait de même avec les classes encodées sous forme one-hot
classes = np.argmax(train_labels, axis=1)

print "Les 10 premières classes"
print classes[:10], "\n"

# On peut comparer tous les indices en même temps grace à numpy
print "Le résultat des 10 premières comparaisons"
print (classes != predictions)[:10], "\n"

# Il reste seulement à calculer la moyenne. Chaque True est évalué à 1 et False à 0
print "La moyenne des comparaisons évaluées à True"
print (classes != predictions).mean(), "\n"

# Et voilà, reste plus qu'à l'afficher sous forme de pourcentage
print "Le taux d'erreur de classification de notre modèle bidon!"
print (classes != predictions).mean() * 100, "%", "\n"
```
