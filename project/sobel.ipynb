```python
import cPickle
import scipy.ndimage as ndimage
from sklearn import linear_model
```

```python
# try with mnist digit 
f = open("mnist.pkl", "rb")
(train_x, train_y), (valid_x, valid_y), (test_x, test_y)= cPickle.load(f)
train_x = np.concatenate((train_x, valid_x))
train_y = np.concatenate((train_y, valid_y))
```

```python
lin_classifier = linear_model.SGDClassifier()
```

```python
def apply_sobel(vector_image):
    """ applies the sobel filter in x and y directions"""
    assert len(vector_image.shape) == 2  # working with an array
    sx = ndimage.sobel(vector_image, axis=0, mode='constant')
    sy = ndimage.sobel(vector_image, axis=1, mode='constant')
    return (sx, sy)

def sobelize(vector_image):
    if (len(vector_image.shape) != 2):
        v_img = np.reshape(vector_image, (28, 28))
    else:
        v_img = vector_image
    sx = ndimage.sobel(v_img, axis=0, mode='constant')
    sy = ndimage.sobel(v_img, axis=1, mode='constant')
    return np.ndarray.flatten(np.hypot(sx, sy))
```

```python
train_x_s = map(sobelize, train_x)
test_x_s = map(sobelize, test_x)
```

```python
lin_classifier.fit(train_x_s, train_y)
```

```python
def accuracy(data, target, classifier):
    return np.sum(np.equal(classifier.predict(data), target)) / float(len(target))
```

```python
print accuracy(train_x_s, train_y, lin_classifier)
print accuracy(test_x_s, test_y, lin_classifier)
```
