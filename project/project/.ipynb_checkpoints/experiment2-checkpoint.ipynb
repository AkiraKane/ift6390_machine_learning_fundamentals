```python
import numpy as np
import cPickle
import sklearn.linear_model
import sklearn.datasets
import sklearn.cross_validation  # we'll use stratified kfold

# get the general utilities
import utility
reload(utility)

# preprocessing functions
import preprocessing
reload(preprocessing)

# performance measure functions
import analysis
reload(analysis)

# better print function
from __future__ import print_function

# plotting functions
import seaborn

# cython performance
import pyximport; pyximport.install(reload_support=True)
```

```python
# GENERAL DATA HANDLING

# load the mnist dataset
mnist = sklearn.datasets.fetch_mldata('MNIST original', data_home='.')
mnist.data = map(lambda x: x / 256., mnist.data)  # bring to float


# preprocessing (deskew and to 8-bit)
preprocess = lambda x: preprocessing.to8bit(preprocessing.deskew(x))
mnist.data = np.reshape(map(preprocess, mnist.data), (len(mnist.data), 784))
```

```python
# generate the data
train_x, test_x, train_y, test_y = \
            sklearn.cross_validation.train_test_split(mnist.data,
                                                      mnist.target,
                                                      train_size=int(6.*len(mnist.target)/7.),
                                                      random_state=1)
train_x = np.reshape(train_x, (len(train_x), 784))
test_x = np.reshape(test_x, (len(test_x), 784))
```

```python
# initialize the l1 and l2 classifiers
penalties = ['l2', 'l1', 'elasticnet']
alphas = [0.0000000000001, 0.0001, 0.001, 0.1, 0.5]
classifiers_l1 = [sklearn.linear_model.SGDClassifier(loss='hinge', penalty='l1', alpha=a) for a in alphas]
classifiers_l2 = [sklearn.linear_model.SGDClassifier(loss='hinge', penalty='l2', alpha=a) for a in alphas]


# initialize the elasticnet classifiers
classifiers_015 = [sklearn.linear_model.SGDClassifier(loss='hinge', penalty='elasticnet', l1_ratio=0.15, alpha=a) for a in alphas]
classifiers_05 = [sklearn.linear_model.SGDClassifier(loss='hinge', penalty='elasticnet', l1_ratio=0.5, alpha=a) for a in alphas]
classifiers_085 = [sklearn.linear_model.SGDClassifier(loss='hinge', penalty='elasticnet', l1_ratio=0.85, alpha=a) for a in alphas]
```

```python
# train all the classifiers
acceptable_classifiers = []

# keep only those above a fixed threshold of accuracy
threshold = 0.85

for classifier in (classifiers_l1 + classifiers_l2 + classifiers_015 + classifiers_05 + classifiers_085):
    classifier.fit(train_x, train_y)
    score = classifier.score(test_x, test_y)
    if score > threshold:
        acceptable_classifiers.append(classifier)

```

```python
# calculate the distances for equal sample of all classes
import pyximport; pyximport.install(reload_support=True)
import distortion_fast
reload(distortion_fast)

results = []

for classifier in acceptable_classifiers:
    %time tmp = distortion_fast.distances_and_class(mnist.data, mnist.target, classifier, 150, 42)
    results.append(tmp)
```

```python
# flatten the data

flat_data = []

for (_, i) in results:
    flat_data.append(np.ndarray.flatten(i))
```

```python
# plot the data

seaborn.set(style="white")

names1 = []
names2 = []

# for the colors
penalties2 = []

data1 = []
data2 = []

scores = []
for (i, classifier) in enumerate(acceptable_classifiers):
    p = classifier.penalty
    score = classifier.score(test_x, test_y)
    a = classifier.alpha
    scores.append(score)
    if p=='l1':
        names2.append(u'{0}\nα ={1}\n% = {2}'.format(p, a, score))
        data2.append(flat_data[i])
        penalties2.append(p)
    elif p=='l2':
        names1.append(u'{0}\nα ={1}\n% = {2}'.format(p, a, score))
        data1.append(flat_data[i])
    else:
        names2.append(u'{0}\nα ={1}\nr = {2}\n% = {3}'.format(p, a, classifier.l1_ratio, score))
        data2.append(flat_data[i])
        penalties2.append(p)

# create the plots
seaborn.set_style("white")

# plot
# l2 penalty
plt.figure(figsize=(10, 6))
plt.xlabel("squared euclidian norm to adversarial example")
plt.ylabel("classifiers and training parameters, n={0}".format(len(flat_data[0])))
seaborn.boxplot(data1, color="skyblue", names=names1, vert=False)
savefig("../draft/figs/experiment_2_l2.eps")
plt.show()

plt.figure(figsize=(10, 13))
plt.xlabel("squared euclidian norm to adversarial example")
plt.ylabel("classifiers and training parameters, n={0}".format(len(flat_data[0])))
seaborn.boxplot(data2, color="skyblue", names=names2, vert=False)
savefig("../draft/figs/sup_1.eps")
plt.show()
```
