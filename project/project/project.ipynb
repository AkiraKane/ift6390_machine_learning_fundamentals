## Data loading

```python
# GENERAL DATA HANDLING

# load the mnist dataset
mnist = sklearn.datasets.fetch_mldata('MNIST original', data_home='.')
mnist.data = map(lambda x: x / 256., mnist.data)  # bring to float


# preprocessing (deskew and to 8-bit)
preprocess = lambda x: preprocessing.to8bit(preprocessing.deskew(x))
mnist.data = np.reshape(map(preprocess, mnist.data), (len(mnist.data), 784))
```

# Experiment 1
verify that the size of the interval on the true value of $\lVert{r} \rVert^2$ is always small

to assess that, we make sure that the relative size of the interval to the distance calculated is small



```python
# generate the data
train_x, test_x, train_y, test_y = \
            sklearn.cross_validation.train_test_split(mnist.data,
                                                      mnist.target,
                                                      train_size=int(6.*len(mnist.target)/7.),
                                                      random_state=1)
train_x = np.reshape(train_x, (len(train_x), 784))
test_x = np.reshape(test_x, (len(test_x), 784))

# train the classifiers
reg = ['l2', 'l1', 'elasticnet']
classifiers = [sklearn.linear_model.SGDClassifier(loss='hinge', penalty=r, alpha=0.001) for r in reg] # initialize the linear svm
map(lambda x: x.fit(train_x, train_y), classifiers)

# check the accuracy of the classifiers
map(lambda x: print("precision = " + str(analysis.get_global_accuracy(test_x, test_y, x))), classifiers)
```

```python
import distortion
reload(distortion)

# we'll use 1000 different images from the training set
size = 200
np.random.seed(42)
indices = np.random.choice(np.arange(0, len(train_y)), size*3)

intervals = []
for classifier in classifiers:
    interval = []
    adequate = 0
    for index in indices:
        original_class = train_y[index]
        digit = train_x[index]
        if classifier.predict(digit) == original_class:
            adequate += 1
            v = distortion.get_squared_norm_adversarial(digit, original_class, classifier)
            interval.append(v)
        if adequate >= size:
            break
    intervals.append(interval)

```

```python
# PLOT
# experiment 1

palette = seaborn.hls_palette(8, l=.3, s=.8)
color_dict = {'l1':palette[1], 'l2':palette[2], 'elasticnet':palette[0]}
seaborn.set(style="white")

interval_rel = []
for i in intervals:
    tmp = []
    for j in i:
        tmp.append(j[1] - j[0])
    interval_rel.append(tmp)


names = [u'{1}\nα ={0}'.format(c.alpha, c.penalty) for c in classifiers]
seaborn.boxplot(interval_rel, color=[color_dict[i] for i in reg], names=names)
#plt.title("interval width of the squared norm for different classifiers, n={0}".format(len(interval_rel[0])))
plt.ylabel("interval width around squared norm")
plt.xlabel("classifier penalty types and weight, n = {0}".format(len(interval_rel[0])))
savefig('../figs/experiment1.eps')
```

This demonstrates that our heuristic is acceptable (the interval is relatively very small to the expected value).

# Experiment 2

We'd like to investigate the relation with various regularization methods and parameters (we always use hinge loss).

```python
# generate the data
train_x, test_x, train_y, test_y = \
            sklearn.cross_validation.train_test_split(mnist.data,
                                                      mnist.target,
                                                      train_size=int(6.*len(mnist.target)/7.),
                                                      random_state=1)
train_x = np.reshape(train_x, (len(train_x), 784))
test_x = np.reshape(test_x, (len(test_x), 784))
```

```python
# initialize the l1 and l2 classifiers
penalties = ['l2', 'l1', 'elasticnet']
alphas = [0.0000000000001, 0.0001, 0.001, 0.1, 0.5]
classifiers_l1 = [sklearn.linear_model.SGDClassifier(loss='hinge', penalty='l1', alpha=a) for a in alphas]
classifiers_l2 = [sklearn.linear_model.SGDClassifier(loss='hinge', penalty='l2', alpha=a) for a in alphas]


# initialize the elasticnet classifiers
classifiers_015 = [sklearn.linear_model.SGDClassifier(loss='hinge', penalty='elasticnet', l1_ratio=0.15, alpha=a) for a in alphas]
classifiers_05 = [sklearn.linear_model.SGDClassifier(loss='hinge', penalty='elasticnet', l1_ratio=0.5, alpha=a) for a in alphas]
classifiers_085 = [sklearn.linear_model.SGDClassifier(loss='hinge', penalty='elasticnet', l1_ratio=0.85, alpha=a) for a in alphas]
```

```python
# train all the classifiers
acceptable_classifiers = []

for classifier in (classifiers_l1 + classifiers_l2 + classifiers_015 + classifiers_05 + classifiers_085):
    # 
    classifier.fit(train_x, train_y)
    score = classifier.score(test_x, test_y)
    if score > 0.85:
        #print("classifier {0} is good".format(i))
        acceptable_classifiers.append(classifier)

```
