```python
# GENERAL DATA HANDLING

# load the mnist dataset
mnist = sklearn.datasets.fetch_mldata('MNIST original', data_home='.')
mnist.data = map(lambda x: x / 256., mnist.data)  # bring to float


# preprocessing (deskew and to 8-bit)
preprocess = lambda x: preprocessing.to8bit(preprocessing.deskew(x))
mnist.data = np.reshape(map(preprocess, mnist.data), (len(mnist.data), 784))
```

```python
x_1, x_2, y_1, y_2 = \
            sklearn.cross_validation.train_test_split(mnist.data,
                                                      mnist.target,
                                                      train_size=30000,
                                                      random_state=42)
x_1 = np.reshape(x_1, (len(x_1), 784))
x_2 = np.reshape(x_2, (len(x_2), 784))
```

```python
classifier1 = sklearn.linear_model.SGDClassifier(loss='hinge', penalty='l2', alpha=0.5)
classifier2 = sklearn.linear_model.SGDClassifier(loss='hinge', penalty='l2', alpha=0.5)

classifier1.fit(x_1, y_1)
classifier2.fit(x_2, y_2)
```

```python
acc1 = analysis.get_global_accuracy(x_2, y_2, classifier1)
acc2 = analysis.get_global_accuracy(x_1, y_1, classifier2)

print("accuracies of the classifiers are {0} and {1}".format(acc1, acc2))
```

```python
# generate some distorted examples
import distortion_fast
reload(distortion_fast)

distorted1 = distortion_fast.distances_and_class(mnist.data, mnist.target, classifier1, 100, 1337)
distorted2 = distortion_fast.distances_and_class(mnist.data, mnist.target, classifier2, 100, 1337)
```

```python
import distortion

def generateImages(distorted_images, classifier, data):
    # gennerate distorted image (x, y)
    height = len(distorted_images[0])*10
    X = numpy.zeros((height, 784))
    Y = numpy.zeros(height, dtype='i')
    counter = 0
    for original_class,c in enumerate(distorted_images):
        for (index, target, squared_norm) in c:
            digit = data[index]
            new_image = distortion.distort_opt_ratio(digit, original_class, target, classifier, get_image=True)
            X[counter] = new_image+digit
            Y[counter] = original_class
            counter += 1
    return (X, Y)
```

```python
# generate distorted images
x_1_adv, y_1_adv = generateImages(distorted1, classifier1, mnist.data)
x_2_adv, y_2_adv = generateImages(distorted2, classifier2, mnist.data)
```

```python
# verify the accuracy on the second classifier
assert(analysis.get_global_accuracy(x_1_adv, y_1_adv, classifier1) == 0.)
assert(analysis.get_global_accuracy(x_2_adv, y_2_adv, classifier2) == 0.)

acc_1_adv = analysis.get_global_accuracy(x_2_adv, y_2_adv, classifier1)
acc_2_adv = analysis.get_global_accuracy(x_1_adv, y_1_adv, classifier2)

class_acc_1_adv = analysis.get_class_performance(x_2_adv, y_2_adv, classifier1)
class_acc_2_adv = analysis.get_class_performance(x_1_adv, y_1_adv, classifier2)
```
