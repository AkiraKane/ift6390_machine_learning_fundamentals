```python
import cPickle
import scipy.ndimage as ndimage
```

```python
# try with mnist digit 
f = open("mnist.pkl", "rb")
(train_x, _), _, _= cPickle.load(f)
a = train_x[0]
a = np.reshape(a, (28, 28))
plt.imshow(a, cmap=plt.cm.gray)
```

```python
sx = ndimage.sobel(a, axis=0, mode='constant')
sy = ndimage.sobel(a, axis=1, mode='constant')
sob = np.hypot(sx, sy)
plt.imshow(sx, cmap=plt.cm.gray)
plt.show()

plt.imshow(sy, cmap=plt.cm.gray)
plt.show()

plt.imshow(sob, cmap=plt.cm.gray)
plt.show()
```
